```python
pob_t_d = (
    pob_t.merge(id_trabajos)
)
```

```python
pob_t_d = pob_t_d.drop(pob_t_d[pob_t_d.edo_conyug == 6].index)
```

```python
pob_t_d['cod']= np.where((pob_t_d['edo_conyug'] == 1) | (pob_t_d['edo_conyug'] == 2) | (pob_t_d['edo_conyug'] == 5), 1, 0)
```

```python
pob_t_d['cod']
```

```python
pob_t_d.head(2)
```

```python
gtd = pob_t_d.groupby(['sinco', 'cod'])
gtd.head()
```

```python
gtd_grupo=gtd.size().unstack()
```

```python
gtd_grupo.sort_values(by=0, ascending=False)
```

```python
gtd1= (
    gtd_grupo.reset_index().
    merge(id_trabajos)
)
gtd1.head()
```

```python
#gtd1=gtd1.rename(columns={'0': 'div','1': 'cas'})
gtd1=gtd1.rename(columns = {0:'div', 1: 'cas'})
gtd1.head()
```

```python
gtd1.columns
```

```python
gtd1['porcentaje']=(gtd1['div']/(gtd1['cas']+gtd1['div']))*100
gtd1.head()
```

```python
gtd2=gtd1.sort_values(by='porcentaje', ascending=False).reset_index()
gtd2.head()
```

```python
gtd2.to_csv('divorcios_p1.csv')
```
